```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')

from sklearn.manifold import TSNE
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

import data.preprocessing as dp
```

```python
# data csv file path
path = '/content/drive/MyDrive/bistelligence/BIS_data/BISTelligen_Project_Data.csv'

# get data
data = dp.GetData(path)
```

```python
noisy_index = data.loc[data['Vibration_RMS2']<0,'Vibration_RMS2'].index
wo_noisy_data = data.drop(noisy_index).copy()
wo_noisy_data = wo_noisy_data.loc[wo_noisy_data.key==4]
#X = wo_noisy_data.loc[:,['Vibration_Peak2','Power2']]
#X = pd.DataFrame(wo_noisy_data['Vibration_Peak2'])
X = wo_noisy_data.drop(['Vibration_RMS2'],axis=1)
y = wo_noisy_data['Vibration_RMS2']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=2023)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_score = model.score(X_train,y_train)
y_score
```

```python
mean_squared_error(y_test,y_pred)
```
